```python
# %%writefile textutils.py
"""
textutils.py - helper functions for string processing (intentionally buggy!)
"""

VOWELS = set("aeiou")

def normalize_text(s: str) -> str:
    """Trim spaces and lowercase the text."""
    return s.strip().lower()

def split_words(s: str) -> list[str]:
    """Split text into words by whitespace."""
    return s.split(" ")

def unique_words(words: list[str]) -> set[str]:
    """Return a set of unique words (case-insensitive)."""
    words = [normalize_text(w) for w in words]
    return set(words)

def count_vowels(s: str) -> int:
    """Count vowels in a string (case-insensitive)."""
    s = s.lower()
    vowels = set("aeiou")
    return sum(1 for ch in s if ch in vowels)
```

```python

students = [
    {
        "name": "  Alice  ",
        "scores": [88, 92, 79],
        "tags": {"new", "excellent"},
        "meta": ("CSE101", 1),
    },
    {
        "name": "bOB",
        "scores": [95, 90, 93, 97],
        "tags": {"scholarship"},
        "meta": ("CSE101", 1),
    },
    {
        "name": "Charlie  ",
        "scores": [70, 85, 83],
        "tags": {"needs-help", "retake?"},
        "meta": ("CSE102", 2),
    },
]

def average(nums: list[float]) -> float:
    if not nums:
        return 0.0
    return sum(nums) / len(nums)

def min_max(nums: list[float]) -> tuple[float, float]:
    """Return (min, max)."""
    return (min(nums), max(nums))

def canonical_name(raw_name: str) -> str:
    return normalize_text(raw_name)

def tags_summary(students_data: list[dict]) -> set[str]:
    all_tags = set()
    for s in students_data:
        all_tags |= s["tags"]
    return all_tags

def student_report(students_data: list[dict]) -> None:
    for i in range(len(students_data) ):
        s = students_data[i]
        name = canonical_name(s["name"])
        scores = s["scores"]
        avg = average(scores)
        hi, lo = min_max(scores)
        if avg == 90:
            category = "Outstanding"
        elif avg >= 80 and avg < 90:
            category = "Strong"
        else:
            category = "Needs Improvement"
        print(f"{name}: avg={avg:.2f}, hi={hi}, lo={lo}, category={category}")

print("=== Quick checks ===")
print("Unique tags:", tags_summary(students))
student_report(students)
```
